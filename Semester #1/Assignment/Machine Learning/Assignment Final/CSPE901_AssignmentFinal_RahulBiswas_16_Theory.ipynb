## 1. Derivation of the Sigmoid Activation Function

Logistic regression models the conditional probability:

$$
P(y = 1 \mid \mathbf{x}) = p
$$

We assume the **log-odds (logit)** is a linear function of the input features:

$$
\log\left(\frac{p}{1 - p}\right) = \mathbf{w}^T \mathbf{x} + b
$$

Exponentiating both sides:

$$
\frac{p}{1 - p} = e^{\mathbf{w}^T \mathbf{x} + b}
$$

Solving for \( p \):

$$
p = (1 - p)e^{\mathbf{w}^T \mathbf{x} + b}
$$

$$
p + pe^{\mathbf{w}^T \mathbf{x} + b} = e^{\mathbf{w}^T \mathbf{x} + b}
$$

$$
p = \frac{e^{\mathbf{w}^T \mathbf{x} + b}}{1 + e^{\mathbf{w}^T \mathbf{x} + b}}
$$

Let:

$$
z = \mathbf{w}^T \mathbf{x} + b
$$

Then:

$$
p = \frac{1}{1 + e^{-z}} = \sigma(z)
$$

## 2. Derivation of Log Loss from Maximum Likelihood Estimation

Assume the target variable $ y \in \{0,1\} $ follows a Bernoulli distribution:

$$
P(y \mid \mathbf{x}) = \hat{y}^y (1 - \hat{y})^{1 - y}
$$

where:

$$
\hat{y} = \sigma(\mathbf{w}^T \mathbf{x} + b)
$$

For a dataset of \( m \) independent observations, the likelihood is:

$$
\mathcal{L} = \prod_{i=1}^{m} \hat{y}_i^{y_i}(1 - \hat{y}_i)^{1 - y_i}
$$

Taking the logarithm:

$$
\log \mathcal{L} =
\sum_{i=1}^{m}
\left[
y_i \log(\hat{y}_i) + (1 - y_i)\log(1 - \hat{y}_i)
\right]
$$

To convert this into a minimization problem, we take the **negative log-likelihood (NLL)**:

$$
\mathcal{L}_{\text{NLL}} =
- \sum_{i=1}^{m}
\left[
y_i \log(\hat{y}_i) + (1 - y_i)\log(1 - \hat{y}_i)
\right]
$$

Averaging over all samples gives the **Binary Cross-Entropy (Log Loss)**:

$$
\mathcal{L}_{\text{BCE}} =
-\frac{1}{m} \sum_{i=1}^{m}
\left[
y_i \log(\hat{y}_i) + (1 - y_i)\log(1 - \hat{y}_i)
\right]
$$


## 3. Why Mean Squared Error is Problematic with Sigmoid Activation

Mean Squared Error (MSE) is defined as:

$$
\mathcal{L}_{\text{MSE}} = \frac{1}{2}(\hat{y} - y)^2
$$

When used with a sigmoid activation:

$$
\hat{y} = \sigma(z)
$$

### Mathematical Issue

The gradient of the sigmoid is:

$$
\sigma'(z) = \sigma(z)(1 - \sigma(z))
$$

When \( z \) is very large or very small:

$$
\sigma'(z) \approx 0
$$


This leads to vanishing gradient problem, where gradients become too small
